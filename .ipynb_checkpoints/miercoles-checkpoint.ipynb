```python
def split_string_by_addition_signs(string):
    new_string = string
    if "+" in string:
        new_string = string.split("+")[0]
    elif "-" in string:
        new_string = string.split("-")[0]
    return new_string
```

```python
df_drop2[columns1] = df_drop2[columns1].applymap(lambda x: split_string_by_addition_signs(x))
```

```python
forwards=['RF', 'ST', 'LW', 'LF', 'RS', 'LS', 'RM', 'LM','RW']
midfielders=['RCM','LCM','LDM','CAM','CDM','LAM','RDM','CM','RAM','CF']
defenders=['RCB','CB','LCB','LB','RB','RWB','LWB']
goalkeepers=['GK']

def pos2(position):
    if position in forwards:
        return 'Forward'
    
    elif position in midfielders:
        return 'Midfielder'
    
    elif position in defenders:
        return 'Defender'
    
    elif position in goalkeepers:
        return 'GK'
    
    else:
        return 'nan'
```
